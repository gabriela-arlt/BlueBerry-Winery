```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine['quality_label'] = red_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
red_wine['quality_label'] = pd.Categorical(red_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
white_wine['quality_label'] = white_wine['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
white_wine['quality_label'] = pd.Categorical(white_wine['quality_label'],
categories=['low', 'medium', 'high'])
```

```python
pd.concat([red_wine[['residual sugar', 'total sulfur dioxide','sulphates', 'alcohol','volatile acidity','quality']].describe(), white_wine[['residual sugar', 'total sulfur dioxide','sulphates', 'alcohol','volatile acidity','quality']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
#sulfates and volatile acidity
```

```python
# AVG of fixed acidiy for red wine is higher than for the white wine/ std big difference

# Fixed acidity: acids are major wine properties and contribute greatly to the wineâ€™s taste. Usually, the total acidity is divided into 
# two groups: the volatile acids and the nonvolatile or fixed acids. Among the fixed acids that you can find in wines are the following:
# tartaric, malic, citric, and succinic. This variable is expressed in g(tartaricacid)/dm3 in the data sets.

pd.concat([red_wine[['fixed acidity']].describe(), white_wine[['fixed acidity']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
# AVG free sulfur dioxide for white wine is two times higher than the red whine/std also higher for white wine
# Free sulfur dioxide: the part of the sulfur dioxide that is added to a wine and that is lost into it is said to be bound, 
# while the active part is said to be free. The winemaker will always try to get the highest proportion of free sulfur to bind. 
# This variable is expressed in mg/dm3 in the data.

pd.concat([red_wine[['free sulfur dioxide']].describe(), white_wine[['free sulfur dioxide']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
# avg cloride is higher for the red wine compared to the white wine also std
# Chlorides can be a significant contributor to saltiness in wine. Here, youâ€™ll see that itâ€™s expressed in g(sodiumchloride)/dm3

pd.concat([red_wine[['chlorides']].describe(), white_wine[['chlorides']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```

```python
# avg residual sugar us 3 times higher for white wine also std
pd.concat([red_wine[['residual sugar']].describe(), white_wine[['residual sugar']].describe()], axis=1, keys=["Red Wines Stats", "White Wines Stats"])
```
