```python
import numpy as np
import pandas as pd

def cap_outliers(wine2, method='iqr', multiplier=1.5):
    """
    Capping outliers in a DataFrame.

    Parameters:
    - df: pandas DataFrame
    - method: str, optional (default='iqr')
        The method for capping outliers. Options: 'iqr' (Interquartile Range)
    - multiplier: float, optional (default=1.5)
        Multiplier for determining the outlier threshold.

    Returns:
    - wine2_capped: pandas DataFrame
        DataFrame with capped outliers.
    """
    wine2_capped= wine2.copy()

    for column in wine2_capped.columns:
        if method == 'iqr':
            Q1 = wine2_capped[column].quantile(0.25)
            Q3 = wine2_capped[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - multiplier * IQR
            upper_bound = Q3 + multiplier * IQR

            # Cap outliers
            wine2_capped[column] = np.clip(wine2_capped[column], lower_bound, upper_bound)

    return wine2_capped

# Example usage:
# Assuming 'your_dataset' is your DataFrame
# capped_df = cap_outliers(your_dataset, method='iqr', multiplier=1.5)
```
