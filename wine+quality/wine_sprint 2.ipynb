```python
import numpy as np
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
import seaborn as sns
```

## Heatmap RED WINE

```python
red_wine_heatmap = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv', sep=';')
white_wine_heatmap = pd.read_csv('winequality-white.csv', sep=';')
```

```python
white_wine_heatmap
```

```python
white_wine_heatmap.drop_duplicates(inplace=True)
red_wine_heatmap.drop_duplicates(inplace=True)
white_wine_heatmap.reset_index(inplace=True)
red_wine_heatmap.reset_index(inplace=True)
red_wine_heatmap.drop(columns='index', inplace=True)
white_wine_heatmap.drop(columns='index', inplace=True)
```

```python
red_wine_heatmap.head()
```

## Lecture

```python
red_wine_heatmap.hist(bins=4, color='red', edgecolor='black', linewidth=3.0, xlabelsize=6, ylabelsize=6, grid=False)

plt.tight_layout(rect=(0, 0,1.2, 1.2))

rt = plt.suptitle('Red Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python
white_wine_heatmap.hist(bins=4, color='#ffeda0', edgecolor='black', linewidth=3.0, xlabelsize=6, ylabelsize=6, grid=False)

plt.tight_layout(rect=(0, 0,1.2, 1.2))

rt = plt.suptitle('White Wine Univariate Plots', x=0.65, y=1.25, fontsize=15)
```

```python
# add column wine type
red_wine_heatmap['wine type']='red'
white_wine_heatmap['wine type']='white'
wine=pd.concat([red_wine_heatmap, white_wine_heatmap], ignore_index=True)
wine
```

```python
# For example, take the feature named residual sugar and plot the distributions across data pertaining to red and white wine samples.

fig, axs= plt.subplots(nrows = 1, ncols= 2, figsize = (10,5))
fig.suptitle('Residual Sugar Content in Wine', fontsize=20, color= '#dd1c77')
sns.histplot(bins=20, x = "residual sugar" , stat='frequency', data = red_wine_heatmap, color='#dd1c77', ax=axs[0])
sns.histplot(bins=20, x = "residual sugar" ,stat='frequency', data = white_wine_heatmap, fill=False, color='black', ax=axs[1])
axs[0].set_ylim(0, 2500)
axs[1].set_ylim(0, 2500)
axs[1].set_title('White Wine')
axs[0].set_title('Red Wine')
# axs[0].set_facecolor("#636363")
# axs[1].set_facecolor("#636363")

```

```python
sns.set_style ('white')
my_colors=['Red','Yellow']

title = "Wine Qiality-Sulphates'"
g=sns.jointplot(x='quality', y='sulphates', data=wine, hue='wine type', kind='kde', palette= my_colors, marginal_ticks=True, ratio=2, marginal_kws={'shade':True})
plt.suptitle(title, y = 1.05)
```

```python
sns.set_style ('white')
title = "Red Wine Qiality-Sulphates'"
g=sns.jointplot(x='quality', y='sulphates', data=red_wine_heatmap, color='red', kind='kde', marginal_ticks=True)
plt.suptitle(title, y = 1.05)
sns.despine(left=False, bottom=False, top=False, right=False)
```

## why?

```python
# we are creating a new column called "quality_label", we define a range and associate that range with a label
red_wine_heatmap['quality label'] = red_wine_heatmap['quality'].apply(lambda value: 'low'
if value <= 5 else 'medium'
if value <= 7 else 'high')

# here we are transforming these labels into categrical data type (specific to pandas) instead of simple string
red_wine_heatmap['quality label'] = pd.Categorical(red_wine_heatmap['quality label'],
categories=['low', 'medium', 'high'])
```

```python
my_colors=['Yellow','Orange', 'Blue']
sns.scatterplot(x='alcohol', y='sulphates',data=red_wine_heatmap, hue='quality label', palette=my_colors,s=200, alpha=0.4)
#style='quality lable' - changes the shapes of the markers
#marker='*', edgecolor='black'
```
